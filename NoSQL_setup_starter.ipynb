5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
# Update the data type of longitude and latitude fields
result = collection.update_many({}, [
    {"$set": {"longitude": {"$toDecimal": "$longitude"}}},
    {"$set": {"latitude": {"$toDecimal": "$latitude"}}}
])

# Print the number of modified documents
print("Modified documents:", result.modified_count)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
collection = db["establishments"]

# Define the non-rating values
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]

# Update RatingValue to integers and set non 1-5 ratings to null
result = collection.update_many(
    {"RatingValue": {"$in": non_ratings}},
    [{"$set": {"RatingValue": {"$toInt": {"$cond": [{"$in": ["$RatingValue", non_ratings]}, None, "$RatingValue"]}}}}]
)

# Print the number of modified documents
print("Modified documents:", result.modified_count)
```

```python
# Check that the coordinates and rating value are now numbers
# Find and display one document from the establishments collection
document = collection.find_one()

# Check the data type of the longitude, latitude, and RatingValue fields
print("Longitude data type:", type(document["longitude"]))
print("Latitude data type:", type(document["latitude"]))
print("RatingValue data type:", type(document["RatingValue"]))
```
